```python
# https://facebook.github.io/prophet/docs/quick_start.html
# m = Prophet(changepoint_prior_scale=0.01)
m = NeuralProphet()
#m.add_seasonality(name='hourly', period=1/24, fourier_order=15)
dp = d2['2024-03-01 06:00:00':'2024-05-16 06:00:00'] 
dp.reset_index(inplace=True)
dp.columns.values[0] = "ds"
dp.columns.values[1] = "y"
dp.fillna(0,inplace=True)
print(dp.head(10))
m.fit(dp, freq="h")

```

```python
future = m.make_future_dataframe(periods=24*60, df=dp)   #,freq='h'
future
forecast = m.predict(future)
forecast.tail(5)
```

```python
fig1 = m.plot(forecast)

forecastp = forecast.set_index("ds")
forecastp = forecastp[["yhat1"]]
figf = px.line(forecastp)    # .loc['2024-03-01':'2024-03-3']
figf.show()
forecastp.head()
```
